```python
# Importar as bibliotecas necessárias
import moviepy.editor as mpe
```

```python
# Carregar o vídeo original
video = mpe.VideoFileClip("video.mp4")
```

```python
# Dividir o vídeo em segmentos de 20 segundos e aplicar inversão alternada
clips = []
for start in range(0, int(video.duration), 20):
    subclip = video.subclip(start, min(start + 20, video.duration))
    if (start // 20) % 2 == 0:
        subclip = subclip.fx(mpe.vfx.mirror_x)  # Inverter horizontalmente
    clips.append(subclip)

# Concatenar os clipes processados
inverted_video = mpe.concatenate_videoclips(clips, method="compose")
```

```python
duration = inverted_video.duration
clips = []

# Processar o vídeo em segmentos de 10 segundos para redução de volume
for i in range(0, int(duration), 10):
    start = i
    end = min(i + 10, duration)
    subclip = inverted_video.subclip(start, end)
    
    # Calcular o fator de volume: começa em 1 e vai até 0 nos últimos 10 segundos
    if start < duration - 10:
        volume_factor = 1 - (start / (duration - 10))
    else:
        volume_factor = 0  # Últimos 10 segundos em silêncio
    
    # Aplicar o volume ao segmento
    subclip = subclip.volumex(volume_factor)
    clips.append(subclip)

# Concatenar todos os segmentos processados
video_with_reduced_volume = mpe.concatenate_videoclips(clips, method="compose")
```

```python
# Cortar o trecho entre os segundos 60 e 80
cut_part = video_with_reduced_volume.subclip(60, min(80, video_with_reduced_volume.duration))

# Inserir o trecho cortado no final do vídeo
final_video = mpe.concatenate_videoclips([video_with_reduced_volume, cut_part], method="compose")
```

```python
# Salvar o vídeo final
final_video.write_videofile("output_gradual_reduction.mp4", codec="libx264", audio_codec="aac")
print("Redução de volume completa.")
```
